# Task1: Merging 12 months sale into single file

```python
import pandas as pd
import os
files = [file for file in os.listdir('Pandas-Data-Science-Tasks-Master/SalesAnalysis/Sales_Data')]
df = pd.DataFrame()
for f in files:
    df_temp = pd.read_csv('Pandas-Data-Science-Tasks-Master/SalesAnalysis/Sales_Data/'+f)
    df = pd.concat([df,df_temp])
# dropNa
df.dropna(inplace=True)
index_toDrop = df[df['Order ID']=='Order ID'].index
df.drop(index=index_toDrop,inplace=True)
df.drop_duplicates(inplace=True)
```

# Task 2: Data Type Check

```python
df['Price Each'] = pd.to_numeric(df['Price Each'])
df['Order Date'] = pd.to_datetime(df['Order Date'])
```

```python
df['Quantity Ordered'] = pd.to_numeric(df['Quantity Ordered'])
```

# Task 3: Split Purchase Address and Preparing data for analysis

```python
# df['post_id'] = df['Purchase Address'].str[-5:]
# df['State Address'] = df['']
df['State'] = df['Purchase Address'].str[-8:-6]
df['State'].value_counts()
```

```python
df = df.sort_values('Order Date').reset_index(drop=True)
df
```

```python
def city_finding(text):
    idx = text.index('St,')
    idx += 4
    city = text[idx:]
    city = city[:city.index(',')]
    return city
    
```

```python
txt = df["Purchase Address"][0]
df['City'] = df['Purchase Address'].apply(city_finding)
df
```

```python
df['Street'] = pd.Series(map(lambda x: x[:x.index(',')],df['Purchase Address']))
df
```

```python
df['post_id'] = df['Purchase Address'].str[-5:]
df
```

```python
df.columns
```

```python
df = df[['Order ID', 'Product', 'Quantity Ordered', 'Price Each', 'Order Date','Street', 'City' ,'State', 'post_id',
       'Purchase Address' ]]
```

# Task 4: Categorize Product

```python
# Check all category
df['Category_check'] = df['Product'].str.contains('Headphones|Charging Cable|Batteries|Monitor|Laptop|Phone|TV|Dryer|Washing Machine',regex=True)
```

```python
df[df['Category_check']==False]
```

```python
df.head()
```

```python
Category = ['Headphones','Charging Cable','Batteries','Monitor','Laptop','Phone','TV','Dryer','Washing Machine']
def product_categorize(txt):
    for cate in Category:
        if cate in txt:
            return cate
df['Product Category'] = df['Product'].apply(product_categorize)
```

```python
del df['Category_check']
```

```python
df.head()
```

```python
df['Month'] = df['Order Date'].dt.to_period('M')
```

```python
df['Month_name'] = df['Order Date'].dt.month_name()
```

```python
df.columns
df = df[['Order ID', 'Product','Product Category' ,'Quantity Ordered', 'Price Each', 'Order Date','Month_name','Month',
       'Street', 'City', 'State', 'post_id', 'Purchase Address']]
df.head()
```
